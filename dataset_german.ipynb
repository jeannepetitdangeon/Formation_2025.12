```python
categorical_sub = ['Sex', 'Housing', 'Saving accounts',
                   'Checking account', 'Purpose']

df = pd.get_dummies(dataset, columns = categorical_sub,
                    drop_first=True).astype(int)
df.head()
```

```python
import statsmodels.api as sm
```

```python
y = df['Credit amount']
X = df.drop(['Credit amount','Unnamed: 0'],axis=1)
X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
print(model.summary())
```

## **ModÃ¨le de machine learning âš“**

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import ExtraTreesRegressor
```

```python
df.head()
```

```python
categorical_sub = ['Sex', 'Housing', 'Saving accounts',
                   'Checking account', 'Purpose']

df2 = pd.get_dummies(data, columns = categorical_sub,
                    drop_first=True).astype(int)

y = df2['Credit amount']
X = df2.drop(['Credit amount','Unnamed: 0'],axis=1)

X=sm.add_constant(X)

imputer = IterativeImputer(
    estimator = ExtraTreesRegressor(n_estimators=100,random_state=42),
                           random_state=42
)

X_imputed = imputer.fit_transform(X)
X_imputed = pd.DataFrame(X_imputed, columns= X.columns)
X_imputed.head()
```

```python
X_imputed = sm.add_constant(X_imputed)
model = sm.OLS(y,X_imputed).fit()
print(model.summary())
```
