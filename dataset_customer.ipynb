# **Analyse des données ⛅**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualise le churn en graphique bâton avec deux couleurs
sns.countplot(data = dataset,x='Churn',fill=True)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
sns.boxplot(y=dataset['Age'],x=dataset['Churn'],
            hue=dataset['Churn'],ax=ax, palette='Set3', legend=False)
ax.tick_params(axis = 'x', rotation = 45)
```

```python
# Visualiser le churn par rapport à l'âge dans un histogramme
sns.kdeplot(data=dataset, x='Age', hue='Churn', fill=True)
```

```python
sns.kdeplot(data=dataset, x='Years', hue='Churn', fill=True)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
sns.boxplot(y=dataset['Years'],x=dataset['Churn'],
            hue=dataset['Churn'],ax=ax, palette='Set3', legend=False)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
sns.boxplot(y=dataset['Total_Purchase'],x=dataset['Churn'],
            hue=dataset['Churn'],ax=ax, palette='Set3', legend=False)
ax.tick_params(axis = 'x', rotation = 45)
```

```python
sns.countplot(data=dataset, x='Churn', hue='Account_Manager')
```

```python
sns.scatterplot(data = dataset, x = 'Num_Sites', y = 'Years')
```

```python
sns.boxplot(data = dataset, x = 'Account_Manager', y = 'Num_Sites')
```
