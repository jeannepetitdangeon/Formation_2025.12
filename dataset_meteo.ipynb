```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Supprimer les colonnes "Unnamed: 0", "Hour", "Minute"
dataset = dataset.drop(["Unnamed: 0", "Hour", "Minute","Day","Year"], axis=1)
dataset.shape
```

```python
# Remplace les points par des _ dans les noms de colonne

dataset.columns = (
    dataset.columns
    .str.replace("..", "_")
    .str.replace(".", "_")
    .str.replace(r'_+$', '', regex=True)
)
dataset.columns
```

```python
import statsmodels.api as sm
```

```python
categorical_sub = ['Month']

df = pd.get_dummies(dataset, columns = categorical_sub,
                    drop_first=True).astype(int)
df.head()
```

```python
y = df['pluie_demain']
X = df.drop(['pluie_demain'],axis=1)
X = sm.add_constant(X)
model = sm.Logit(y,X).fit()
print(model.summary())
```

```python
pvalues_sorted = model.pvalues.sort_values().apply(lambda x: f"{x:.6f}")
print(pvalues_sorted)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=41)

X_train = sm.add_constant(x_train)
model_train = sm.Logit(y_train,X_train).fit()
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred = (y_pred > 0.5).astype(int)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}\n")
```

```python
comparaison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred, 'Match': y_test == y_pred})
similarity_percentage = int((comparaison['Match'].sum() / len(comparaison)) * 100)
print(f"Pourcentage de similarit: {similarity_percentage}%")
```
